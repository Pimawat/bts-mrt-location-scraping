```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }

# LOOP 1-2529
url = "https://en.wikipedia.org/wiki/Sukhumvit_line"
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
#print(soup.prettify())
```

```python
container = soup.find_all(class_="wikitable")
info = container[0]
```

```python
info.find_all('tr')[0]
```

```python
#3-50
info.find_all('tr')[3]
```

```python
info.find_all('tr')[3].find_all('td')[1]
```

```python
info.find_all('tr')[3].find_all('td')[1].a.get('href')
```

```python
info.find_all('tr')[3].find_all('td')[1].a.get('title')
```

```python
info.find_all('tr')[3].find_all('td')[1].a.get_text()
```

```python
url_name = []
full_name = []
name = []

for i in range(3,51):
    url_name.append(info.find_all('tr')[i].find_all('td')[1].a.get('href'))
    full_name.append(info.find_all('tr')[i].find_all('td')[1].a.get('title'))
    name.append(info.find_all('tr')[i].find_all('td')[1].a.get_text())
    
    
```

```python
data = {
    'url_name' : url_name,
    'full_name' : full_name,
    'name' : name
}

data_df = pd.DataFrame(data)
data_df
```

```python
url = "https://en.wikipedia.org/wiki/11th_Infantry_Regiment_BTS_station"
req = requests.get(url, headers)
soup = BeautifulSoup(req.content, 'html.parser')
```

```python
container = soup.find_all(class_="geo-dec")
info = container[0]
```

```python
container
```

```python
info.get_text()
```

```python
location = []
link_ls = []
link_no_data = []

for link in url_name:
    url = "https://en.wikipedia.org"+link
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    
    container = soup.find_all(class_="geo-dec")
    if len(container)>0 :
        info = container[0]
        location.append(info.get_text())
        link_ls.append(link)
    else:
        location.append('')
        link_ls.append(link)
```

```python
location[1].split(' ')
```

```python
float(location[1].split(' ')[0].replace('°N',''))
```

```python
float(location[1].split(' ')[1].replace('°E',''))
```

```python
lat = []
long = []

for i in location :
    print(i)
    if i == '':
        lat.append(0)
        long.append(0)
    else:
        lat.append(float(i.split(' ')[0].replace('°N','')))
        long.append(float(i.split(' ')[1].replace('°E','')))
```

```python
data = {
    'url_name' : url_name,
    'full_name' : full_name,
    'name' : name,
    'lat' : lat,
    'long' : long
}

data_df = pd.DataFrame(data)
data_df
```

```python
data_df.to_csv('bts_sk_line.csv')
```
